```python
import numpy as np
import matplotlib.pyplot as plt

# x = np.linspace(0, 10, 100)
x = np.arange(0, 10, 0.1)
y = np.sin(x)
plt.plot(x, y, label='sin(x)', linestyle=':')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Wave')
plt.legend()
# plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x = np.linspace(0, 10, 100)
x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)', linestyle='-.')
plt.plot(x, y2, label='cos(x)', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend()
plt.show()
```
