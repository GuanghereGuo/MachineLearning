```python
#测试基本功能
import seaborn as sns
import matplotlib.pyplot as plt

# sns.set_theme(style="whitegrid", palette="pastel", context="talk")
sns.set_theme(style="whitegrid", palette="pastel", context="paper")
# sns.set_theme(style="whitegrid", palette="pastel", context="notebook")

products = ["Product A", "Product B", "Product C", "Product D"]
sales = [120, 210, 150, 180]

sns.barplot(x=products, y=sales, label="test01")

# 添加标签和标题
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product Sales by Category")
plt.legend()

# 显示图表
plt.show()
```

```python
# 测试pandas兼容性
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 创建一个示例 DataFrame
data = {
    '年龄': np.random.randint(20, 60, 100),
    '收入': np.random.randint(3000, 15000, 100) + np.random.randn(100) * 1000,
    '性别': np.random.choice(['男', '女'], 100),
    '产品购买量': np.random.randint(1, 10, 100),
    '满意度': np.random.randint(1, 6, 100)
}
df = pd.DataFrame(data)

print("示例 DataFrame 头部：")
print(df.head())
```

```python

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='收入', bins=20, kde=True)
plt.title('收入分布直方图')
plt.xlabel('收入 (元)')
plt.ylabel('计数')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='收入', fill=True)
plt.title('收入分布核密度估计图')
plt.xlabel('收入 (元)')
plt.ylabel('密度')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='满意度', hue='性别')
plt.title('不同性别对产品满意度的计数')
plt.xlabel('满意度等级')
plt.ylabel('计数')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.regplot(data=df, x='年龄', y='收入', scatter_kws={'alpha':0.6})
plt.title('年龄与收入的线性回归关系')
plt.xlabel('年龄')
plt.ylabel('收入 (元)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x='性别', y='收入', inner='quartile') # inner='quartile' 显示四分位数
plt.title('不同性别的收入分布小提琴图')
plt.xlabel('性别')
plt.ylabel('收入 (元)')
plt.show()
```
