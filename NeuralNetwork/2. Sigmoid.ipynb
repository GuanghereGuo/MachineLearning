```python
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x):
    return 1 / (1 + np.exp(-x))  # broadcasting


def step_function(x):
    return np.where(x > 0, 1, 0)


def ReLU(x):
    return np.maximum(0, x)



# def step_function(x):
#     return np.array(x > 0, dtype=np.int32)

x = np.arange(-10, 10, 0.1)
y1 = sigmoid(x)
y2 = step_function(x)
y3 = ReLU(x)

plt.plot(x, y1, label="sigmoid", linestyle='--')
plt.plot(x, y2, label="step function")
plt.plot(x, y3, label="ReLU", linestyle=':')
plt.title("Sigmoid and Step Function")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

```
